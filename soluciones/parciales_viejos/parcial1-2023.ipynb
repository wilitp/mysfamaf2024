```python
from functools import reduce
import math
from typing import List, Tuple
from random import random


print("2")

def jugadaA(u, v):

    return u > 0.5 and v < 0.5

def jugadaB(u, v):

    return u < 0.5 and v > 0.5

def sim(n):
    count = 0

    for _ in range(0, n):
        u ,v = random(), random()
        if jugadaA(u,v):
            count += 1
        elif not jugadaB(u, v):
            u ,v = random(), random()
            if jugadaA(u,v):
                count += 1
    return count/n

# print(sim(1_000_000))

def montecarlo_multi(g, intervals: List[Tuple[float, float]], n: int):
    """
    Calcula una integral mÃºltiple
    """

    assert g.__code__.co_argcount == len(intervals)

    # obtener a partir de los x_i los u_i a partir de cambios de variable
    # el cambio de variable depende de si los limites es a, b ambos finitos
    # o si son a = 0 y b = infinito
    def aux(): 

        factors = []
        args = []

        for a, b in intervals:
            x = random()
            u = (x * (b - a) + a) if b is not math.inf else 1/x - 1
            p = (b - a) if b is not math.inf else 1/x**2
            args.append(u)
            factors.append(p)
        
        return [reversed(args), factors]

    # generador de samples para obtener el promedio
    def gen_func():
        for _ in range(0, n):

            args, prods = aux()

            prod = reduce(lambda x, y : x*y, prods)

            yield g(*args) * prod

    gen = gen_func()

    # devolver el promedio de la muestra generada
    return (sum(gen))/n

# print(montecarlo_multi(lambda x : x / (x - math.exp(x)), [(-3, 3)], 1_000_000))

# print(montecarlo_multi(lambda x : x**3 * math.exp(-x**3), [(0, math.inf)], 1_000_000) + montecarlo_multi(lambda x : x**3 * math.exp(-x**3), [(-1, 0)], 1_000_000))

def sim(n):
    count = 0
    for _ in range(0, n):
        u, v = random(), random()

        if max(u, v) > 0.6:
            count += 1
    return count/n

# print(sim(1_000_000))

print(montecarlo_multi(lambda x, y : 1 - math.exp(-(x+y)), [(0,1), (0,1)], 2_000_000))

```
